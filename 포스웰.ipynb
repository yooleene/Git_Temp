```python
from selenium import webdriver

driver = webdriver.Chrome('chromedriver')

driver.implicitly_wait(10)

area_code = 'D31'
year = '2022'
month = '07'
day = '15'

url = 'https://www.poswel.co.kr/fmenu/todaymenu.php?area_code={}&nyear={}&nmonth={}&reqday={}'.format(area_code,year,month,day)

driver.get(url)
driver.maximize_window()
```

```python
menu = driver.find_element_by_xpath('//*[@id="today_menu_list01"]/div[2]/div')
print(menu.text)
```

```python
menu = driver.find_element_by_id('today_menu_list')
menu.text
```

```python
course = driver.find_elements_by_css_selector('.blue.bold700.font17')
won = driver.find_elements_by_css_selector('.blue.blod700.font20')
menu = driver.find_elements_by_css_selector('.list_menu_explain.font11.lineheight18')

# course = driver.find_elements_by_css_selector('span.blue.bold700.font17')
# won = driver.find_elements_by_css_selector('span.blue.bold700.font20')
# menu = driver.find_elements_by_css_selector('div.today_menu_list01_box02 > div')
```

```python
cc = []
ww = []
mm = []

for c in course:
      cc.append(c.text)
      
for w in won:
      ww.append(w.text)
      
for m in menu:
      mm.append(m.text)
```
