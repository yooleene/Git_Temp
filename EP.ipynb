```python
import pyautogui as pg
from selenium import webdriver
import sys
import clipboard as cb
```

```python
my_id = 'yooleenes'
my_pw = pg.password(title='EP 패스워드', text='패스워드를 입력하세요')
```

```python

url = 'http://swp.posco.net'

driver = webdriver.Chrome('chromedriver')
driver.get(url)

driver.implicitly_wait(10)
driver.maximize_window()

driver.find_element_by_xpath('//*[@id="username"]').send_keys(my_id)
driver.find_element_by_xpath('//*[@id="password"]').send_keys(my_pw)
driver.find_element_by_xpath('//*[@id="loginsubmit"]').click()

pg.sleep(5)
try:
      alert = driver.switch_to.alert
      alert.accept()
except:
      print('기존 로그인창 없음')

```

```python
pg.sleep(5)#15초 정도

for i in range(-1, len(driver.window_handles)): #-1,0,1,2
      success_check = False
      driver.switch_to.window(driver.window_handles[i])
      this_page_title = driver.title
      if 'EP(Enterprise Portal)' in this_page_title:
            try:
                  print('ECM 아이콘 클릭중')
                  driver.find_element_by_xpath('//*[@id="4"]/a[1]/span[1]/img').click()
            except:
                  print('ECM 아이콘 클릭 실패')
            else:
                  print('ECM 아이콘 클릭 완료')
                  success_check = True
                  break
if success_check is False:
      print('진행 오류 발생')
      sys.exit()
```
