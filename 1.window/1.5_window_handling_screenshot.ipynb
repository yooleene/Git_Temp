# 라이브러리 불러오기

```python
import pyautogui
```

# 스크린샷의 캡처

```python
image01 = pyautogui.screenshot  #스크린 전체 화면을 이미지 객체로 전환

#스크린샷을 이미지 저장할 때는 pillow 라이브러리가 설치되어 있어야 함(아나콘다에는 자동 살치되어 있음)
pyautogui.screenshot('image02.png')  #스크린 전체 화면을 image02.png 현재 폴더에 파일로 저장, 전체 경로 지정 가능
pyautogui.screenshot('image02.png', region=(0, 0, 500, 600))
#특정영역 지정 : region 설정(x좌표 상단 좌측, y좌표 상단 좌측, x좌표 하단 우측, y좌표 하단 우측)
```

# 이미지를 찾아 좌표 반환

```python
pyautogui.locateOnScreen('win_b.png') #win_b.png 이미지를 찾아 좌표를 반환 
```

```python
pyautogui.locateCenterOnScreen('win_b.png') #win_b.png 이미지를 찾아 정중앙의 좌표를 반환 
```

```python
pyautogui.locateCenterOnScreen('win_b.png', region=(0,1000, 200,1080)) #특정 범위에서 이미지를 검색(속도와 정확도를 올림)
```

# 이미지를 찾아 반환된 좌표를 클릭

```python
win_b = pyautogui.locateCenterOnScreen('win_b.png', region=(0,1000, 200,1080))
pyautogui.click(win_b)
```
