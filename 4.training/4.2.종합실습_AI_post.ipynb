## 3.2 데이터 분석(계산)

```python
import win32com.client
import string
import pyautogui

excel = win32com.client.Dispatch('Excel.Application')
excel.Visible = True

wb = excel.Workbooks.Open(file_name)
ws = wb.Activesheet

pyautogui.sleep(3)
ws.Range('1:3').Delete() #상단의 정보가 있는 행 삭제
ws.Range('C:C').Delete() #화씨 온도 열 삭제
ws.Range('A:A').ColumnWidth = 15 #날자 열 너비 변경

#평균 구하기
i = 1
cell_value = ws.Cells(1,1).Value #초기 값으로 아무 셀 값이나 설정

while(cell_value != None):
      i+=1
      cell_value = ws.Cells(i,1).Value
max_row = i-1

ws.Cells(max_row+1,1).Value = '평균'

abc = string.ascii_lowercase
for col in range(2,5):
      ws.Cells(max_row+1,col).Formula = '=round(average({}2:{}{}),2)'.format(abc[col-1], abc[col-1], max_row)

ws.Columns('A:D').HorizontalAlignment  = 3 #가운데 정렬

wb.Save()
pyautogui.sleep(3)
excel.Quit()
```

# 4.보고서 작성

## 4.1 보고서 작성에 필요한 데이터 가져오기

```python
import win32com.client
import pyautogui

excel = win32com.client.Dispatch('Excel.Application')
excel.Visible = True

wb = excel.Workbooks.Open(file_name)
ws = wb.Activesheet

pyautogui.sleep(3)
title_row = ws.Range('B1:D1').Value #튜플로 반환
aver_range = 'B' + str(max_row+1) + ':D' + str(max_row+1)
aver_row = ws.Range(aver_range).Value #튜플로 반환

print(title_row[0]) #첫번째 튜플 사용
print(aver_row[0])
excel.Quit()
```

## 4.2 보고서 PPT 작성(제목, 표, 차트)

```python
from pptx import Presentation
from pptx.util import Inches

prs = Presentation() # 파워포인트 객체 선언

#표 만들기
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN

title_slide_layout = prs.slide_layouts[5] #제목만 슬라이드
slide = prs.slides.add_slide(title_slide_layout) #슬라이드를 파워포인트 객체에 추가

shapes = slide.shapes
shapes.title.text = '열지수 분석 보고서'

rows = 2
cols = 3
left = Inches(0.5)
top = Inches(1.5)
width = Inches(9)
height = Inches(1)

table = shapes.add_table(rows, cols, left, top, width, height).table #add_table() 함수를 호출하여 표를 추가합니다. 
col_name = title_row[0]


#컬럼 크기
j = 0

for col in col_name:
      table.columns[j].width = Inches(3)
      j+= 1

# 표 헤드
j = 0
for col in col_name:
      table.cell(0, j).text = col_name[j]
      #서식
      p = table.cell(0, j).text_frame.paragraphs[0]
      p.font.bold = True
      p.font.size = Pt(26)
      p.font.name = '견고딕'
      p.font.color.rgb = RGBColor(255, 255, 255)
      p.alignment = PP_ALIGN.CENTER

      j+= 1

# 표 값
j = 0
for value in(aver_row[0]):
      table.cell(1, j).text = str(format(value,',')) #천단위 콤마 형식, 표에는 string
      #서식
      p = table.cell(1, j).text_frame.paragraphs[0]
      p.font.size = Pt(24)
      p.alignment = PP_ALIGN.CENTER

      j+= 1

#차트 이미지 가져오기
top = Inches(2.8)
left = Inches(0.3)
height = Inches(4.5)
file_name = 'weather_heat_index_' + now_date + '.png'
pic = slide.shapes.add_picture(file_name, left, top, height=height)

prs.save('weather_heat_index_' + now_date + '.pptx')
```
