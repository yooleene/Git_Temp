```python
ws.Columns('A:E').HorizontalAlignment = 3
```

```python
ws.Range('A:E').AutoFilter(Field=3, Criteria1 = '>0.3' )
ws.Range('A:E').AutoFilter(Field=4, Criteria1 = '>0.5' )
```

```python
i = 1

cell_value = ws.Cells(i,1).Value

while cell_value != None:
      i+=1
      cell_value = ws.Cells(i,1).Value

max_row = i-1

print(max_row)
```

```python
ws.Cells(max_row+1, 1).Value = '합계'
```

```python
import string

abc = string.ascii_lowercase

for col in range(2,6):
      ws.Cells(max_row+1, col).Formula = '=sum({}2:{}{})'.format(abc[col-1], abc[col-1], max_row )
      
```

```python
j = 1

cell_value = ws.Cells(1,j).Value

while cell_value != None:
      j+=1
      cell_value = ws.Cells(1,j).Value

max_col = j-1

print(max_col)
```

```python
ws.Cells(1, max_col+1).Value = '평균'
```

```python
import string

abc = string.ascii_lowercase

for row in range(2,max_row+2):
      ws.Cells(row, max_col+1).Formula = '=average(B{}:{}{})'.format(row, abc[max_col-1], row )
 
```

```python
ws.Range(ws.Cells(1,1), ws.Cells(12,6)).BorderAround(ColorIndex=1, Weight=2, LineStyle=1 )
```

```python
for row in range(1, max_row+2):
      for col in range(1, max_col+2):
            ws.Cells(row,col).BorderAround(ColorIndex=3, Weight=2, LineStyle=1)
```
